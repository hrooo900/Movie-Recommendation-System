```python
# Keeping => genres,id,keywords, title, overview, cast, crew

moovies = moovies[['id','genres','keywords','title','overview','cast','crew']]
moovies.head(1)
```

```python
moovies.isna().sum()
# Removing columns with no overwies

moovies.dropna(inplace=True)
```

```python
# Checking Duplicate Rows
moovies.duplicated().sum()
```

```python
moovies.loc[4,'genres']          
# str, hence need to be converted into a list
```

```python
import ast
ast.literal_eval(moovies.loc[3,'genres'])
```

```python
# Defining a function to convert this object into list of useful keywords only

def convert_genres(obj):
    obj = ast.literal_eval(obj)
    l=[]
    for i in obj:
        l.append(i['name'])
    return l

convert_genres(moovies.loc[3,'genres'])
```

```python
# Preview on entire genres columns
moovies['genres'].apply(convert_genres)
```

```python
moovies['genres'] = moovies['genres'].apply(convert_genres)     # Don't run this line again
moovies.head(2)
```

```python
moovies.loc[10,'keywords'] 
# convert_genres can also be used for this column change
```

```python
moovies['keywords']= moovies['keywords'].apply(convert_genres)
moovies.head(1)
```

```python
moovies.loc[12,'cast']
# Only First 3 dictionaries to be taken into account
```

```python
def convert_cast(oj):
    oj = ast.literal_eval(oj)
    l = []
    for i in oj[:3]:
        l.append(i['name'])
        
    return l
        
convert_cast(moovies['cast'][88])
```

```python
moovies['cast'] = moovies['cast'].apply(convert_cast)
moovies.head(1)
```

```python
moovies.loc[234,'crew']
```

```python
def convert_crew(oj):
    oj = ast.literal_eval(oj)
    
    l=[]
    for i in oj:
        if i['job']=="Director":
            l.append(i['name'])
    return l
convert_crew(moovies.loc[234,'crew'])
```

```python
moovies['crew'] = moovies['crew'].apply(convert_crew)
moovies.head(1)
```

```python
moovies['overview'] = moovies['overview'].apply(lambda x: x.split())
moovies.head(1)
```

```python
# Zoe Saldana ===> ZoeSaldana              Science Fiction ==>  ScienceFiction

# While creating tags, system should point unique tags

for col in ['genres','keywords','cast','crew']:
    moovies[col] = moovies[col].apply(lambda x: [i.replace(" ","") for i in x])
    
moovies.head(1)
```

```python
moovies['tags'] = moovies['genres'] + moovies['keywords']+ moovies['overview'] + moovies['cast'] + moovies['crew']
moovies.head(1)
```

```python
moovies[moovies['id']==47327]
```

```python
df = moovies[['id','title','tags']]
df.head(1)
```

```python
# Converting tags list into str

df['tags'] = df['tags'].apply(lambda x : " ".join(x))
df.head(1)
```

```python
df['tags'][0]
```

```python
# Lowering all tags
df['tags'] = df['tags'].apply(lambda x : x.lower())
df.head(1)
```

### Text Vectorizer

```python
from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer(max_features = 5000, stop_words= 'english')

vectors = cv.fit_transform(df['tags']).toarray()
# Takes time

vectors
```

```python
cv.get_feature_names_out()[30:100]

# can be observed that words are repeating like (agent,agents), (allow,allows), (answer,answers) 
```

```python
# Repairing above Fault

import nltk
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()

# Checking stem method of PorterStemmer
for i in ['creation','creativity','created','create']:
    out = ps.stem(i)
    print(out)
```

```python
# Creating a helper function
def cutting(text):
    y = []
    
    for i in text.split():
        word = ps.stem(i)
        y.append(word)
    return ' '.join(y)
    
cutting(df['tags'][0])
```

```python
df['tags'] = df['tags'].apply(cutting)
df.head(1)
```

```python
# Rechecking Text Vectorization
from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer(max_features = 5000, stop_words= 'english')

vectors = cv.fit_transform(df['tags']).toarray()
vectors
```

```python
cv.get_feature_names_out()[300:400]
```

```python
vectors.shape
```

```python
# Now calculating distance between each pair of vectors using cosine similarity

# cosine similarity is inversly proportional to cosine distance

from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vectors)              # shape = (4806, 4806)
```

```python
def recommend(movie):
    movie_index = df[df['title'] == movie].index[0]
    rec_movies = sorted(list(enumerate(similarity[movie_index])),reverse = True, key = lambda x: x[1])[1:6]
    
    for i in rec_movies:
        print(df['title'][i[0]])
        
        
recommend('Transformers')
```
